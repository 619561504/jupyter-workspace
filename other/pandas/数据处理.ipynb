```python
def split_output(filename, output_prefix, base_path):
    datas = pd.read_table(os.path.join(base_path, filename), engine="python" )
    datas["imei"].to_csv(os.path.join(base_path, output_prefix + "_全部.csv"), index=False)
    datas[datas["is_new"] == 0]["imei"].to_csv(os.path.join(base_path, output_prefix + "_老用户.csv"), index=False)
    datas[datas["is_new"] == 1]["imei"].to_csv(os.path.join(base_path, output_prefix + "_新用户.csv"), index=False)
```
